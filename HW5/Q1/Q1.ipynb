#### Part b

```python
dataset = pd.read_csv('Hackathon_Working_Data.csv')
dataset.isna().sum()
# No nan values
```

#### Part c

```python
dataset['ID'] = dataset[['STORECODE', 'BILL_ID']].sum(1)
```

#### Part d

```python
cart_df = dataset.groupby(by=['ID', 'GRP'], as_index=False).sum().fillna(0)
cart_df
```

#### Part e

```python
cart_df.loc[cart_df['QTY'] < 1, 'QTY'], cart_df.loc[cart_df['QTY'] >= 1, 'QTY'] = 0, 1
cart_df = cart_df.dropna()
cart_df
```

#### Part f

```python
transactions = cart_df.groupby('ID', as_index=True)['GRP'].apply(lambda z: list(z))
transactions_df = pd.DataFrame(
    index=transactions.index,
    columns=np.unique(np.concatenate(transactions.to_list()))
)
```

```python
for index, columns in transactions.items():
    transactions_df.loc[
        index,
        columns
    ] = True
    transactions_df.loc[index, :].fillna(False, inplace=True)
transaction_df = transactions_df.astype(bool)
```

```python
transaction_df
```

```python
from mlxtend.frequent_patterns import apriori
frequent_items_df = apriori(transaction_df, min_support=0.01, use_colnames=True)
```

#### Part g

```python
from mlxtend.frequent_patterns import association_rules
assoc_rules_df = association_rules(frequent_items_df, metric='lift')
assoc_rules_df
```

#### Part h

There is a rule that associates `(TOOTH PASTES) -> (WASHING POWDERS/LIQUIDS)`. This means that in the bills that had tooth paste in them there is a high chance that there is also washing powder/liquids in them.

The `leverage` metric calculates the difference of the frequency of the antecedents and consequent happening together and the frequency of each one happening independently.

$levarage(A→C) = support(A→C) − support(A) × support(C), range: [−1,1]$

The `conviction` metric computes the dependency of the consequent to the antecedent.

$conviction(A→C)= \frac{1−support(C)}{1−confidence(A→C)}, range: [0,∞]$
