```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import jenkspy
```

```python
dataset = pd.read_csv('machines.csv', index_col=0)
dataset.columns = ['Vendor_name', 'model', 'MCYT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX', 'PRP', 'ERP']
```

### 6.a

```python
pd.cut(dataset['CACH'], bins=3).value_counts().plot.bar()
```

We divide the entire range of values by 3 and then put every value in its respective created range.

### 6.b

```python
pd.qcut(dataset['CACH'], q=3).value_counts().plot.bar()
```

In this method, instead of the dividing the range of values by 3, we divide the values in a way that each bin have approximately the same number of values as other bins.

### 6.c

```python
pd.cut(dataset['CACH'], bins=jenkspy.jenks_breaks(dataset['CACH'], nb_class=3)).value_counts().plot.bar()
```

In this method, we divide the entire range into 3 number of bins and put our values in it in a way that in each bin, the variance is minimum. in other words, we try to find 3 clusters in our data and make them our bins.
