```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('train.csv', index_col='PassengerId')
```

### 7.a

```python
dataset.describe()
```

### 7.b

```python
dataset.select_dtypes('number').mean().sort_values().plot.bar()
```

### 7.c

```python
dataset.plot.box(subplots=True, figsize=(15, 5))
```

### 7.d

```python
fig, ax = plt.subplots()
sns.heatmap(data=dataset.corr(), annot=True, ax=ax)
fig.set_size_inches((12, 7))
```

We can see that Parch and SibSp are the most correlated features.

### 7.e

```python
fig, ax = plt.subplots()
sns.heatmap(data=dataset.isna(), ax=ax)
fig.set_size_inches((12, 7))
```

```python
(dataset.isna().sum() / dataset.shape[0])*100
```

```python
print(f'Column with the most missing data: {(dataset.isna().sum() / dataset.shape[0]).idxmax()}')
```
