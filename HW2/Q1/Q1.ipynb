### Question 1

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('credit cards.csv')
```

### Part a

```python
fig, ax = plt.subplots()
sns.heatmap(data=dataset.corr(), ax=ax, annot=True)
fig.set_size_inches((20, 12))
```

We can see that columns `bill_amt1` to `bill_amt6` have high correlation with each other. Also, columns `pay_1` to `pay_6` have high correlation with each other. both cases are valid and normal because they show a families bills and expenses over time which we expect to change very little over time and be correlated. other than these, there is no highly noticeable correlation between other columns. 

### Part b

```python
highly_correlated = [f'bill_amt{i}' for i in range(1, 7)]
bill_amt_df = dataset[highly_correlated]
sns.pairplot(data=bill_amt_df)
```

In the figures matrix, on the diagonal line we can see the distribution of each column. in the other cells, we can see pair-wise points of all pairs of columns. if the resulting shape of the points is closer to a straight line, we can say that the respecting pair of columns have a higher correlation.

### Part c

```python
spearman_corr_df = dataset.corr(method='spearman')
fig, ax = plt.subplots()
sns.heatmap(data=spearman_corr_df[['default']], ax=ax, annot=True, yticklabels=spearman_corr_df.index)
fig.set_size_inches((16, 10))
```

The `pay_1` to `pay_6` have the most correlation with the `default` column and the `pay_1` column has the highest one. Using the spearman correlation's definition, we can assume that the `default` value is a monotonic function of the `pay_x` columns.

### Part d

```python
kendall_corr_df = dataset.corr(method='kendall')
fig, ax = plt.subplots()
sns.heatmap(data=kendall_corr_df[['default']], ax=ax, annot=True, yticklabels=kendall_corr_df.index)
fig.set_size_inches((16, 10))
```

Same as part c, the `pay_x` columns have the highest correlation with the `default` column. the result is same as part c.

### Part e

```python
fig, ax = plt.subplots()
sns.heatmap(data=dataset.corr(method='spearman'), ax=ax, annot=True)
fig.set_size_inches((20, 12))
```

Generally, both methods show high correlations between the same columns, but because the pearson method requires linear correlation, it is more strict, but the spearman method only requires monotonic correlation(both values must increase or decrease together for this method to show high correlation). as the result of this less strict requirement of correlation, spearman method values are typically larger.
